#Data Preparation

1.import library

```python
import pandas as pd
import numpy as np
import math
from datetime import datetime, timedelta
```

2.อ่านไฟล์

```python
student = pd.read_csv('student.csv')
```

3.แสดงข้อมูลในตาราง

```python
student
```

4.ทำสำเนา

```python
student_copy = student.copy()
```

5.ตรวจดูค่าว่าง

```python
student_copy.isnull().sum()
```

6. ลบแถวทั้งหมดที่มีค่า NaN (ค่าหาย) ออกจาก DataFrame

```python
student_copy.dropna(subset=['ชื่อ'], inplace=True)
student_copy
```

7. หาค่าเฉลี่ยอายุนักเรียนทั้งหมดเพื่อนำมาเติมค่าลงในช่องว่าง และวนลูปว่ามีค่าในหลักไหนของ 'อายุ' ที่มากกว่า 30 แล้วแทนด้วยค่าเฉลี่ย

```python
mean = round(student_copy['อายุ'].mean())
student_copy['อายุ'].fillna(mean, inplace=True)
for i, age in student_copy['อายุ'].items():
    if age  > 30:
        student_copy.at[i, 'อายุ'] = mean

student_copy
```

8. จับคู่เพศที่มีความหมายเดียวกัน เพื่อให้ข้อมูลมีความสม่ำเสมอ

```python
gender_mapping = {
    'ชาย': 'ชาย',
    'หญิง': 'หญิง',
    'F': 'หญิง',
    'M': 'ชาย',
    'ญ': 'หญิง',
    'ช': 'ชาย',
    'Female': 'หญิง',
    'Male' : 'ชาย'
}
```

9. นำค่าที่จับคู่มาปรับใช้กับตาราง ส่วนค่าว่างให้แทนด้วย ไม่ระบุ

```python
student_copy['เพศ']=student_copy['เพศ'].map(gender_mapping)
student_copy['เพศ'].fillna('ไม่ระบุ', inplace=True)

student_copy
```

10. จับคู่คณะที่มีความหมายเดียวกัน ให้เท่ากับค่าๆหนึ่ง เพื่อให้ข้อมูลมีความสม่ำเสมอ

```python
faculty_mapping = {
    'วิทย์': 'วิทยาศาสตร์',
    'คณะวิทยาศาสตร์' : 'วิทยาศาสตร์',
    'Science': 'วิทยาศาสตร์',
    'Sci' : 'วิทยาศาสตร์',
    'วิทยาศาสตร์': 'วิทยาศาสตร์',
    'ศิลปศาสตร์': 'ศิลปศาสตร์',
    'Arts': 'ศิลปศาสตร์',
    'Liberal Arts': 'ศิลปศาสตร์',
    'ศิลป์' : 'ศิลปศาสตร์',
    'อุตสาหกรรม': 'วิศวกรรมศาสตร์',
    'วิศวกรรมศาสตร์': 'วิศวกรรมศาสตร์',
    'วิศวกรรม': 'วิศวกรรมศาสตร์',
    'วิศวกรรมอุตสาหกรรม': 'วิศวกรรมศาสตร์',
    'วิศวะ': 'วิศวกรรมศาสตร์',
    'วิศวฯ': 'วิศวกรรมศาสตร์',
    'Engineering': 'วิศวกรรมศาสตร์'
}
```

11. นำค่าที่จับคู่มาปรับใช้กับตาราง ส่วนค่าว่างให้แทนด้วย ไม่ระบุ

```python
student_copy['คณะ'] = student_copy['คณะ'].map(faculty_mapping)
student_copy['คณะ'].fillna('ไม่ระบุ', inplace=True)

student_copy
```

12.หาค่าเฉลี่ยเกรดนักเรียนทั้งหมดในตารางเพื่อเอามาแทนในค่าว่าง วนลูปว่ามีค่าในหลักไหนของ 'GPA' ที่มากกว่า 4.00 และมีค่าน้อยกว่า 0.00 แล้วแทนด้วยค่าเฉลี่ย

```python
GPAmean = round(student_copy['GPA'].mean(), 2)
student_copy['GPA'].fillna(GPAmean, inplace=True)
for i, GPA in student_copy['GPA'].items():
    if GPA  > 4.00:
        student_copy.at[i, 'GPA'] = GPAmean

    if GPA < 0.00:
        student_copy.at[i, 'GPA'] = GPAmean

student_copy
```

13. หาค่าเฉลี่ยของน้ำหนัก, ส่วนสูงและชั่วโมงเรียนแล้วแทนในค่าว่าง

```python
heighmean = round(student_copy['ส่วนสูง_cm'].mean())
student_copy['ส่วนสูง_cm'].fillna(heighmean, inplace=True)

Weighmean = round(student_copy['น้ำหนัก_kg'].mean())
student_copy['น้ำหนัก_kg'].fillna(Weighmean, inplace=True)

hourmean = round(student_copy['ชั่วโมงเรียน_ต่อสัปดาห์'].mean())
student_copy['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(hourmean, inplace=True)

student_copy
```

3.สร้างตารางแยกจากของเก่าเพื่อให้แสดงผลไม่สับสน

```python
student_health = pd.DataFrame({
    'น้ำหนัก_kg': student_copy['น้ำหนัก_kg'],
    'ส่วนสูง_cm': student_copy['ส่วนสูง_cm']
})
```

4.สร้าง feature ให้ช่วยคิดคำนวน BMI เพื่อนำมาจัดกลุ่ม

```python
student_health['BMI'] = student_health['น้ำหนัก_kg'] / (student_health['ส่วนสูง_cm']/100)**2
student_health['สุขภาพ'] = pd.cut(student_health['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ต่ำกว่าเกณฑ์', 'สมส่วน', 'เกินมาตรฐาน', 'เป็นโรคอ้วน'])

student_health
```

5.สร้างตารางแยกจากของเก่าเพื่อให้แสดงผลไม่สับสน

```python
student_grade = pd.DataFrame(student_copy['GPA'])
```

6.สร้าง feature ให้ช่วยจัดกลุ่มเกรดเฉลี่ย

```python
student_grade['เกรด'] = pd.cut(student_grade['GPA'],
                                   bins=[0, 0.99, 1.29 ,1.59, 2.29, 2.59, 3.29, 3.59, 4.00],
                                   labels=['F','D','D+', 'C', 'C+', 'B','B+', 'A'])
student_grade
```

7.สร้างตารางแยกจากของเก่าเพื่อให้แสดงผลไม่สับสน

```python
student_date = pd.DataFrame(student_copy['วันที่ลงทะเบียน'])
```

8.สร้าง Feature จากวันเวลาโดยให้ช่วยดูให้ว่าวันที่นักเรียนลงทะเบียนคือวันอะไรในสัปดาห์

```python
student_date['วันที่ลงทะเบียน'] = pd.to_datetime(student_date['วันที่ลงทะเบียน'])
student_date['วันในสัปดาห์'] = student_date['วันที่ลงทะเบียน'].dt.day_name()
student_date
```
